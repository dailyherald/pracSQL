```python
# create the supervisor_salaries table and insert only a few columns

sql = """
    CREATE TABLE supervisor_salaries (
        town varchar(30),
        county varchar(30),
        supervisor varchar(30),
        start_date date,
        salary money,
        benefits money
    )
"""

cur.execute(sql) 

# now we open it and write to the table
# http://initd.org/psycopg/docs/cursor.html

with open('supervisor_salaries.csv', 'r') as f:
    # Notice that we don't need the `csv` module.
    next(f)  # Skip the header row.
    cur.copy_from(f, 'supervisor_salaries', sep=',', columns=('town', 'supervisor', 'salary') )
engine.commit() 

sql = "SELECT * FROM supervisor_salaries"

df = pd.read_sql(sql, engine)

df.head(6)
```

```python
# now we're going to try a temporary table
# first we delete the info from the table
sql = """
    DELETE FROM supervisor_salaries
"""

cur.execute(sql)

sql = "SELECT * FROM supervisor_salaries"

df = pd.read_sql(sql, engine)

df.head()
```

```python
# create the temporary table

sql = """
    CREATE TEMPORARY TABLE supervisor_salaries_temp 
    (LIKE supervisor_salaries)
"""
cur.execute(sql)
```

```python
# now we open it and write to the temp table

with open('supervisor_salaries.csv', 'r') as f:
    next(f)  # Skip the header row.
    cur.copy_from(f, 'supervisor_salaries_temp', sep=',', columns=('town', 'supervisor', 'salary') )
engine.commit() 
```

```python
# here's where we insert the data from the temp table
# and also assign a value to the county column
# that's in the temp table

sql = """
    INSERT INTO supervisor_salaries (town, county, supervisor, salary)
    SELECT town, 'Some County', supervisor, salary
    FROM supervisor_salaries_temp
"""
cur.execute(sql)
```

```python
# Drop the temp table

sql = """
    DROP TABLE supervisor_salaries_temp
"""
cur.execute(sql)
```
